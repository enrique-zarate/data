# Cleaning

```python
# Funcion de filtrado de lineas que no tienen todos los datos
# Sirve para sacar las lineas totalizadoras
def get_empty_rows(dataset):
    # Nested 'for' loop and unique value list of 'set'
    return list(set([i for i, s in dataset.iterrows() for v in s.values if str(v) == 'nan']))
```

# Cuarto PDF

```python
tables = camelot.read_pdf("pdfs/pdf4.pdf", pages='1-end') 
year = "2018"


# Declarar cabecera
# header = df.iloc[0]
header = ['NRO'
          ,'RUC'
          ,'NOMBRE O RAZON SOCIAL'
          ,'APORTE TOTAL(mm)'
        ]

# create a empty dataframe
df_contenedor = pd.DataFrame(columns=header)

# iterar por cada tabla del PDF
for table in tables:
    df = table.df
    # remove the first row with loc
    df = df.loc[1:]
    df.columns = header
    df['ANHO'] = year


    # remove rows with empty string based on NRO column with loc
    df = df.loc[df['NRO'] != '']
    # df = df.loc[df['NRO'].notnull()]
    # df = df.dropna(subset=['Nro'])

    # concatenate the dataframe into a container dataframe
    df_contenedor = pd.concat([df_contenedor, df], ignore_index=True)

    list_of_empty_rows = get_empty_rows(df)

    # remove rows of list_of_empty_rows off of df_contenedor
    df_contenedor = df_contenedor.drop(list_of_empty_rows)
```

## Limpiar y unir a DF madre

```python
# replace '.' with '' and parse APORTE TOTAL(mm) as int
df_contenedor['APORTE TOTAL(mm)'] = df_contenedor['APORTE TOTAL(mm)'].str.replace('.', '')
df_contenedor['APORTE TOTAL(mm)'] = df_contenedor['APORTE TOTAL(mm)'].astype(int)
# filter only the rows with 'TABACALERA DEL ESTE SA' or 'BANCO ITAU PARAGUAY S.A' in NOMBRE O RAZON SOCIAL of df_madre
df_contenedor = df_contenedor.loc[(df_contenedor['NOMBRE O RAZON SOCIAL'] == 'TABACALERA DEL ESTE SA') | (df_contenedor['NOMBRE O RAZON SOCIAL'] == 'BANCO ITAU PARAGUAY S.A')]
# parse ANHO as int
df_contenedor['ANHO'] = df_contenedor['ANHO'].astype(int)
# Concat to df_madre and sort by ANHO column
df_madre = pd.concat([df_madre, df_contenedor], ignore_index=True)
df_madre = df_madre.sort_values(by=['ANHO'], ascending=True)
```

# PDFs 5 al 9
### Años 2017 al 2013

```python
year = "2017"       # inicia el anho

for i in range(5, 10):
    tables = camelot.read_pdf("pdfs/pdf" + str(i) + ".pdf", pages='1-end')

    # empty dataframe
    df_contenedor = pd.DataFrame(columns=header)

    for table in tables:
        df = table.df
        # remove the first row with loc
        df = df.loc[1:]
        df.columns = header
        df['ANHO'] = year

        # remove rows with empty string based on NRO column with loc
        df = df.loc[df['NRO'] != '']

        # concatenate the dataframe into a container dataframe
        df_contenedor = pd.concat([df_contenedor, df], ignore_index=True)
        # get empty rows
        list_of_empty_rows = get_empty_rows(df)

        # remove rows of list_of_empty_rows off of df_contenedor
        df_contenedor = df_contenedor.drop(list_of_empty_rows)
    
    # replace '.' with '' and parse APORTE TOTAL(mm) as int
    df_contenedor['APORTE TOTAL(mm)'] = df_contenedor['APORTE TOTAL(mm)'].str.replace('.', '')
    df_contenedor['APORTE TOTAL(mm)'] = df_contenedor['APORTE TOTAL(mm)'].astype(int)
    # filter only the rows with 'TABACALERA DEL ESTE SA' or 'BANCO ITAU PARAGUAY S.A' in NOMBRE O RAZON SOCIAL of df_madre
    # df_contenedor = df_contenedor.loc[(df_contenedor['NOMBRE O RAZON SOCIAL'] == 'TABACALERA DEL ESTE SA') | (df_contenedor['NOMBRE O RAZON SOCIAL'] == 'BANCO ITAU PARAGUAY S.A') | (df_contenedor['NOMBRE O RAZON SOCIAL'] == 'PARAGUAY REFRESCOS SA')]
    # parse ANHO as int
    df_contenedor['ANHO'] = df_contenedor['ANHO'].astype(int)
    # Concat to df_madre and sort by ANHO column
    df_madre = pd.concat([df_madre, df_contenedor], ignore_index=True)
    df_madre = df_madre.sort_values(by=['ANHO'], ascending=True)

    print(f'''
    Procesando ... \n
    Archivo: pdfs/pdf{i}.pdf\n
    Anho: {year}
    ''')
    

    year = int(year) - 1
    year = str(year)
    
```
